```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://msbooks.pk/"
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html,"html.parser")
```

```python
olevel_notes = soup.find("li",id="menu-item-73364")
olevel_notes_href = olevel_notes.find("a")["href"]
olevel_notes_href
```

```python
olevel_notes_response = requests.get(olevel_notes_href)
olevel_notes_html = olevel_notes_response.text
olevel_notes_soup = BeautifulSoup(olevel_notes_html,"html.parser")
```

```python
notes = olevel_notes_soup.find_all("li",class_="product")
olevel_note_data = []
for note in notes:
    note_href = note.find("div",class_="woocommerce-loop-product__header").find("a")["href"]
    note_response = requests.get(note_href)
    note_html = note_response.text
    note_soup = BeautifulSoup(note_html,"html.parser")

    n = note_soup.find("div",class_="summary entry-summary")

    title = n.find("h1", class_="product_title entry-title").text
    price = n.find("p",class_="price").text
    price = int(price.replace(",","")[2:])

    attributes = n.find("div",class_="msbooks-attributes").find_all("p")
    missing_count = 0

    grade = attributes[0].text
    if "Grade: " in grade:
        grade = grade.replace("Grade: ","")
    else:
        grade = None
        missing_count += 1
    grade = grade.replace("LEVEL", "Level").replace("Levels", "Level")

    author = attributes[1-missing_count].text
    if "Author: " in author:
        author = author.replace("Author: ","")
    else:
        author = None
        missing_count +=1

    binding = attributes[2-missing_count].text
    if "Binding: " not in binding:
        missing_count += 1

    year = attributes[3-missing_count].text
    if "Year/Edition: " in year:
        year = year.replace("Year/Edition: ","")
    else:
        year = None
        missing_count += 1

    size = attributes[4-missing_count].text
    if "Size: " not in size:
        missing_count += 1
        
    code = attributes[5-missing_count].text
    if "Syllabus: " in code:
        code = code.replace("Syllabus: ","")
    else:
        code = None
        missing_count +=1

    type = attributes[6-missing_count].text
    if "Type: " in type:
        type = type.replace("Type: ","")
    else:
        type = None
        missing_count += 1
    type = type.replace("O Level","")

    olevel_note_data.append({
        "Title": title,
        "Subject code": code,
        "Author": author,
        "Grade": grade,
        "Type": type,
        "Edition": year,
        "Price": price
    })
olevel_note_data
```

```python
import pandas as pd
olevel_notes_df = pd.DataFrame(olevel_note_data)
olevel_notes_df
```

```python
olevel_yearly = soup.find("li",id="menu-item-73366")
olevel_yearly_href = olevel_yearly.find("a")["href"]
olevel_yearly_href
```

```python
olevel_yearly_response = requests.get(olevel_yearly_href)
olevel_yearly_html = olevel_yearly_response.text
olevel_yearly_soup = BeautifulSoup(olevel_yearly_html,"html.parser")
```

```python
olevel_yearly_data = []
while True:
    yearlys = olevel_yearly_soup.find_all("li",class_="product")
    print ("Products in page:",len(yearlys))
    for yearly in yearlys:
        yearly_href = yearly.find("div",class_="woocommerce-loop-product__header").find("a")["href"]
        yearly_response = requests.get(yearly_href)
        yearly_html = yearly_response.text
        yearly_soup = BeautifulSoup(yearly_html,"html.parser")

        y = yearly_soup.find("div",class_="summary entry-summary")

        title = y.find("h1", class_="product_title entry-title").text
        price = y.find("p",class_="price").text
        price = int(price.replace(",","")[2:])

        attributes = y.find("div",class_="msbooks-attributes").find_all("p")
        missing_count = 0

        grade = attributes[0].text
        if "Grade: " in grade:
            grade = grade.replace("Grade: ","")
        else:
            grade = None
            missing_count += 1
        grade = grade.replace("LEVEL", "Level").replace("Levels", "Level")

        author = attributes[1-missing_count].text
        if "Author: " in author:
            author = author.replace("Author: ","")
        else:
            author = None
            missing_count +=1

        binding = attributes[2-missing_count].text
        if "Binding: " not in binding:
            missing_count += 1

        year = attributes[3-missing_count].text
        if "Year/Edition: " in year:
            year = year.replace("Year/Edition: ","")
        else:
            year = None
            missing_count += 1

        size = attributes[4-missing_count].text
        if "Size: " not in size:
            missing_count += 1
            
        code = attributes[5-missing_count].text
        if "Syllabus: " in code:
            code = code.replace("Syllabus: ","")
        else:
            code = None
            missing_count +=1

        type = attributes[6-missing_count].text
        if "Type: " in type:
            type = type.replace("Type: ","")
        else:
            type = None
            missing_count += 1
        type = type.replace(" with Mark Scheme","").replace(" with Marking Schemes","")

        olevel_yearly_data.append({
            "Title": title,
            "Subject code": code,
            "Author": author,
            "Grade": grade,
            "Type": type,
            "Edition": year,
            "Price": price
        })
    next_btn = olevel_yearly_soup.find("a",class_="next")
    if not next_btn:
        break
    else:
        next_url = next_btn["href"]
        next_response = requests.get(next_url)
        next_html = next_response.text
        olevel_yearly_soup = BeautifulSoup(next_html,"html.parser")

olevel_yearly_data
```

```python
olevel_yearly_df = pd.DataFrame(olevel_yearly_data)
olevel_yearly_df
```
