```python
def send_alert(message):
    data = {"content": message}

    response = requests.post(WEBHOOK_URL, json=data)

    if response.status_code == 204:
        print("Alert sent successfully.")
    else:
        print("Failed to send alert.")
        print(response.status_code)

old_df = pd.read_csv("Books_cleaned.csv")
new_df = pd.read_csv("new_Books_cleaned.csv")
```

```python
old_titles = set(old_df["Title"])
new_titles = set(new_df["Title"])

added_books = new_titles - old_titles

# added_books = set()
# added_books.add("TEST BOOK FOR API VALIDATION")

if added_books:
    if len(added_books)>1:
        message = f"{len(added_books)} new books detected:\n" + "\n".join(added_books)
    else:
        message = f"{len(added_books)} new book detected:\n" + "\n".join(added_books)
    send_alert(message)
else:
    print("No new books found.")
```

```python
old_prices = dict(zip(old_df["Title"], old_df["Price"]))
new_prices = dict(zip(new_df["Title"], new_df["Price"]))

price_changes = []

for title in old_prices:
    if title in new_prices:
        if old_prices[title] != new_prices[title]:
            price_changes.append(
                f"{title}: Rs.{old_prices[title]} → Rs.{new_prices[title]}"
            )

if price_changes:
    message = "Price changes detected:\n" + "\n".join(price_changes)
    send_alert(message)
else:
    print("No price changes found.")
```
